```python
def get_channel_details(youtube, channel_id):
    # Initialize an empty list to store channel data
    all_data = []

    # Make the API request
    request = youtube.channels().list(
        part='snippet,contentDetails,statistics',
        id=','.join(channel_id)
    )
    response = request.execute()

    # Iterate over the items in the response and extract channel details
    for item in response.get('items', []):
        snippet = item.get('snippet', {})
        statistics = item.get('statistics', {})
        content_details = item.get('contentDetails', {})

        # Create a dictionary to store channel details
        data = {
            'Channel_Name': snippet.get('title', ''),
            'Subscribers': statistics.get('subscriberCount', ''),
            'Views': statistics.get('viewCount', ''),  # Note the change in case
            'Total_videos': statistics.get('videoCount', ''),  # Note the change in case
            'Playlist_id': content_details.get('relatedPlaylists', {}).get('uploads', '')
        }

        # Append the channel data to the list
        all_data.append(data)

    return all_data
```

```python
def get_video_ids(youtube, playlist_id):
    # Make the API request to retrieve video IDs from a playlist
    request = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=playlist_id,  # Use 'playlistId' with a capital 'I'
        maxResults=50
    )
    response = request.execute()

    video_ids = []

    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])

    next_page_token = response.get('nextPageToken')
    next_pages = True

    while next_pages:
        if next_page_token is None:
            next_pages = False
        else:
            request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=playlist_id,  # Use 'playlistId' with a capital 'I'
                maxResults=50,
                pageToken=next_page_token
            )
            response = request.execute()

            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId'])

            next_page_token = response.get('nextPageToken')

    return video_ids
```

```python
def get_video_details(youtube, video_ids):
    all_video_stats = []

    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part='snippet,statistics',
            id=','.join(video_ids[i:i + 50])
        )

        response = request.execute()

        for video in response['items']:
            video_stats = {
                'Title': video['snippet']['title'],
                'Published_date': video['snippet']['publishedAt'],
                'views': video['statistics'].get('viewCount', 0),
                'Likes': video['statistics'].get('likeCount', 0),
                'Dislikes': video['statistics'].get('dislikeCount', 0),
                'Comments': video['statistics'].get('commentCount', 0)
            }

            all_video_stats.append(video_stats)

    return all_video_stats
```
